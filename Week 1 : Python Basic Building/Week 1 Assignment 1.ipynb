**Q3. Explain the use of the following operators using an example:
(i) /
(ii) %
(iii) //
(iv) (**)   **

Explanation -

/ operator is used for normal division

// operator is used for floor division

% operator is used to get the modulus (remainder)

** operator is used for exponential operation ( raised to power )

```python
A = 10
B = 3

# (i) /
i = A/B   # division
print(i)

# (ii) %
ii = A%B  # modulus
print(ii)

# (iii) //
iii = A//B  #Floor division
print(iii)

# (iv) **
iv = A**B # Exponential
print(iv)
```

**Q5. Using a while loop, verify if the number A is purely divisible by number B and if so then how many times it can be divisible.**

```python
A = 100
B = 5

divisions_count = 0

while A % B == 0:
    A = A // B
    divisions_count += 1

if divisions_count > 0:
    print(f"100 was divisible by 5 by {divisions_count} times")
else:
    print(f"{A} is not divisible by {B}")
```

**Q6. Create a list containing 25 int type data. Using for loop and if-else condition print if the element is divisible by 3 or not.**

```python
int_list = [i for i in range(1, 26)]

for num in int_list:
    if num % 3 == 0:
        print(f"{num} is divisible by 3")
    else:
        print(f"{num} is not divisible by 3")
```

**Q7. What do you understand about mutable and immutable data types? Give examples for both showing this property.**

Immutable Data Types:
Immutable data types are those whose state cannot be modified after creation. Any operation that appears to modify the data actually creates a new instance with the modified value, leaving the original data unchanged.

```python
#EXAMPLE - STRING

string = "Hello"
string[0] = 'h'  # This will raise an error because strings are immutable

#EXAMPLE - TUPLES

tuple = (1, 2, 3)
tuple[0] = 4  # This will raise an error because tuples are immutable
```

Mutable Data Types:
Mutable data types, on the other hand, allow modifications to their contents after creation. This means that the state of the object can be changed without creating a new instance.

```python
#EXAMPLE - LISTS
my_list = [1, 2, 3]
my_list[0] = 4  # This modifies the existing list
print(my_list)

#EXAMPLE - DICTIONARIES
my_dict = {'a': 1, 'b': 2}
my_dict['c'] = 3  # This adds a new key-value pair to the dictionary
print(my_dict)
```
